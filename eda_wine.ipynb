```python
for i in numcol:
    data=wi[i]
    q1=np.percentile(data,25)
    q3=np.percentile(data,75)
    iqr=q3-q1
    lb=q1-(1.5*iqr)
    ub=q3+(1.5*iqr)
    m=wi[i].median()
    con=(wi[i]<lb) | (wi[i]>ub)
    tr=m
    fa=wi[i]
    l=np.where(con,tr,fa)
    wi[i]=l
```
